```python
import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('downloads/noguchi02.jpg')
print(image)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
plt.show()
```

```python
print(gray_image)
```

```python
plt.imshow(image)
plt.show()
```

```python
average_color_per_row = np.average(image, axis=0)
average_color = np.average(average_color_per_row, axis=0)
average_color = np.uint8(average_color)
print(average_color)
```

```python
print(average_color_per_row)
```

```python
print(np.average(average_color_per_row, axis=0))
```

```python
average_color_image = np.array([[average_color]*100]*100, np.uint8)
plt.imshow(average_color_image)
```
